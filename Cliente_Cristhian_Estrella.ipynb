```python
class Cliente:
    def __init__(self, nombre, apellido, cedula, email, vip, direccion, telefono, celular, edad):
        self.__nombre = nombre
        self.__apellido = apellido
        self.cedula = cedula          # usa setter
        self.__email = email
        self.__vip = vip
        self.__direccion = direccion
        self.__telefono = telefono
        self.__celular = celular
        self.edad = edad              # usa setter

    # Nombre
    @property
    def nombre(self):
        return self.__nombre

    @nombre.setter
    def nombre(self, valor):
        self.__nombre = valor

    # Apellido
    @property
    def apellido(self):
        return self.__apellido

    @apellido.setter
    def apellido(self, valor):
        self.__apellido = valor

    # Cedula (validar 10 dígitos)
    @property
    def cedula(self):
        return self.__cedula

    @cedula.setter
    def cedula(self, valor):
        if valor.isdigit() and len(valor) == 10:
            self.__cedula = valor
        else:
            print("Cédula inválida: debe tener exactamente 10 dígitos")

    # Email
    @property
    def email(self):
        return self.__email

    @email.setter
    def email(self, valor):
        self.__email = valor

    # VIP
    @property
    def vip(self):
        return self.__vip

    @vip.setter
    def vip(self, valor):
        self.__vip = valor

    # Dirección
    @property
    def direccion(self):
        return self.__direccion

    @direccion.setter
    def direccion(self, valor):
        self.__direccion = valor

    # Teléfono
    @property
    def telefono(self):
        return self.__telefono

    @telefono.setter
    def telefono(self, valor):
        self.__telefono = valor

    # Celular
    @property
    def celular(self):
        return self.__celular

    @celular.setter
    def celular(self, valor):
        self.__celular = valor

    # Edad (mayor de 18)
    @property
    def edad(self):
        return self.__edad

    @edad.setter
    def edad(self, valor):
        if valor >= 18:
            self.__edad = valor
        else:
            print("Edad inválida: debe ser mayor o igual a 18")

    def mostrar_info(self):
        print(f"Nombre: {self.nombre} {self.apellido}")
        print(f"Cédula: {self.cedula}")
        print(f"Email: {self.email}")
        print(f"VIP: {self.vip}")
        print(f"Dirección: {self.direccion}")
        print(f"Teléfono: {self.telefono}")
        print(f"Celular: {self.celular}")
        print(f"Edad: {self.edad}")


# Ejemplo de uso
cliente1 = Cliente("Juan", "Perez", "1234567890", "juan@email.com", True,
                   "Guayaquil", "042345678", "0991234567", 25)

cliente1.mostrar_info()
```
